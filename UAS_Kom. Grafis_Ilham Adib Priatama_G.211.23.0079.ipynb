```python
#Import Module
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
#---------A---------

df = pd.read_csv('stock.csv')

# Membuat grafik histogram 1
plt.figure(figsize=(12,6))
plt.hist(df['Purchase Price'], weights=df['Shares'], bins=20, color='tomato', edgecolor='black')

plt.title('Histogram Purchase Price vs Shares')
plt.xlabel('Purchase Price')
plt.ylabel('Shares')
plt.show()

#____________________________________
# Membuat grafik histogram 2
plt.figure(figsize=(12,6))
plt.hist(df['Current Price'], weights=df['Shares'], bins=20, color='tomato', edgecolor='black')

plt.title('Histogram Current Price vs Shares')
plt.xlabel('Current Price')
plt.ylabel('Shares')
plt.show()

#_____________________________________
# Membuat grafik histogram 3
plt.figure(figsize=(13,6))
plt.hist(df['Gain/Loss'], weights=df['Shares'], bins=20, color='tomato', edgecolor='black')

plt.title('Histogram Gain/Loss vs Shares')
plt.xlabel('Gain/Loss')
plt.ylabel('Shares')
plt.show()
```

```python
#-------B-----------

#sortir "Dividend/share" berdasarkan 5 nilai yang tertinggi
top_lima_dividend = df.sort_values(by='Dividend/share', ascending=False).head(5)

#grafik 1
plt.figure(figsize=(12,6))
plt.bar(top_lima_dividend['Stock Name'], top_lima_dividend['Dividend/share'], color='orange', edgecolor='black')
plt.title('Top 5 Dividend/share berdasarkan Stock Name')
plt.xlabel('Stock Name')
plt.ylabel('Dividend/share')
plt.xticks(rotation=45)
plt.show()

#grafik 2
plt.figure(figsize=(12,6))
plt.bar(top_lima_dividend['Symbol'], top_lima_dividend['Dividend/share'], color='orange', edgecolor='black')
plt.title('Top 5 Dividend/share berdasarkan Symbol')
plt.xlabel('Symbol')
plt.ylabel('Dividend/share')
plt.xticks(rotation=45)
plt.show()
```

```python
#------------C-----------

#sortir data "Market Value" berdasarkan 5 nilai yang tertinggi
top_lima_market_value = df.sort_values(by='Market Value', ascending=False).head(5)

#grafik 1
plt.figure(figsize=(12,6))
plt.bar(top_lima_market_value['Stock Name'], top_lima_market_value['Market Value'], color='cyan', edgecolor='black')
plt.title('Top 5 Market Value berdasarkan Stock Name')
plt.xlabel('Stock Name')
plt.ylabel('Market Value')
plt.xticks(rotation=45)
plt.show()

#grafik 2
plt.figure(figsize=(12,6))
plt.bar(top_lima_market_value['Symbol'], top_lima_market_value['Market Value'], color='cyan', edgecolor='black')
plt.title('5 Tertinggi Market Value berdasarkan Symbol')
plt.xlabel('Symbol')
plt.ylabel('Market Value')
plt.xticks(rotation=45)
plt.show()
```

```python
#-----------C-------------------

#sortir data "Annual Yield" berdasarkan 7 nilai tertinggi tertinggi
top_tujuh_annual_yield = df.sort_values(by='Annual Yield', ascending=False).head(7)

#grafik 1
plt.figure(figsize=(12,6))
plt.bar(top_tujuh_annual_yield['Stock Name'], top_tujuh_annual_yield['Annual Yield'], color='pink', edgecolor='black')
plt.title('Top 7 Annual Yield berdasarkan Stock Name')
plt.xlabel('Stock Name')
plt.ylabel('Annual Yield')
plt.xticks(rotation=45)
plt.show()

#grafik 2
plt.figure(figsize=(12,6))
plt.bar(top_tujuh_annual_yield['Symbol'], top_tujuh_annual_yield['Annual Yield'], color='pink', edgecolor='black')
plt.title('Top 7 Annual Yield berdasarkan Symbol')
plt.xlabel('Symbol')
plt.ylabel('Annual Yield')
plt.xticks(rotation=45)
plt.show()
```
